```python
import sys
sys.path.append('/content/pose')
from tf_pose import common
from tf_pose.estimator import TfPoseEstimator
from tf_pose.networks import get_graph_path, model_wh
```

```python
# Load Pose Model
model='jaad'
pose = TfPoseEstimator(get_graph_path(model), target_size=(224, 224))
```

```python
def get_human_pose(image, resize_to_default = True, showBG = True):

    humans = pose.inference(image, resize_to_default=resize_to_default, upsample_size=4.0)

    if showBG == False:
        image = np.zeros(image.shape)

    image = TfPoseEstimator.draw_humans(image, humans, imgcopy=False)

    return image
```

```python
from xml.etree import ElementTree
import csv
import pandas as pd
import numpy as np
import os
```

```python
files = [f'video_{i:04}' for i in range(1,347)]

# CREATE CSV FILE
csvfile = open("annotation.csv",'w',encoding='utf-8')
csvfile_writer = csv.writer(csvfile)

# ADD THE HEADER TO CSV FILE
csvfile_writer.writerow(["file","time","weather","location","label","frame","keyframe","occluded","outside","xbr","xtl","ybr","ytl",
                         "id","look","reaction","action","cross","hand_gesture","occlusion","nod"])
for filename in files:
  # PARSE XML
  xml = ElementTree.parse(os.path.join(annotation_path,filename+'.xml'))
  attrib = xml.find('meta').find('task').find('video_attributes')
  time_of_day, weather, location = [at.text for at in attrib]
  # FOR EACH PEDESTRIAN
  for track in xml.findall("track"):
    if(track.get('label')=='pedestrian'):
      for box in track:
        frame = box.attrib['frame']
        keyframe = box.attrib['keyframe']
        occluded = box.attrib['occluded']
        outside = box.attrib['outside']
        xbr = int(float(box.attrib['xbr']))
        xtl = int(float(box.attrib['xtl']))
        ybr = int(float(box.attrib['ybr']))
        ytl = int(float(box.attrib['ytl']))
        id = box[0].text
        look = box[2].text
        reaction = box[3].text
        action = box[4].text
        cross = box[5].text
        hand_gesture = box[6].text
        occlusion = box[7].text
        nod = box[8].text
        csv_line = [filename,time_of_day, weather, location, 'pedestrian',frame,keyframe,occluded,outside,xbr,xtl,ybr,ytl,
                    id,look,reaction,action,cross,hand_gesture,occlusion,nod]
        csvfile_writer.writerow(csv_line)
csvfile.close()
```
